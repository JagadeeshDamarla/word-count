```python
import pandas as pd
from collections import Counter
```

```python
# reading the file

file_name = "InputFile.txt"

# dictionary for holding the words
words_list_master = []
words_dict = {}

try:
    #creating file object and reading the file
    with open(file_name) as file:
        for line in file:
            # converting all the words to lower case
            line = str(line).lower()
            words_list = line.split()
            # print(words_list)
            words_list_master.extend(words_list)
        
    # print(words_list_master)
    # adding counter
    words_counter = Counter(words_list_master)
    # print(words_counter)
             
    # # witing the dictionary into a dataframe
    
    
    words_df = pd.DataFrame(words_counter.items(), columns = ["word","frequency"])
    words_df_sorted = words_df.sort_values(by="frequency", ascending = False)
    print(words_df_sorted)
    
                  
except Exception as e:
    print("Exception occured while reading the file "+str(e))
```

```python
### Version 1 of code

# reading the file

file_name = "InputFile.txt"

# dictionary for holding the words
words_dict = {}

try:
    #creating file object and reading the file
    with open(file_name) as file:
        for line in file:
            # converting all the words to lower case
            line = str(line).lower()
            words_list = line.split()

            # skipping the words where empty list is present
            if len(words_list) != 0:
                for word in words_list:
                    # checking if the word is already in the dictionary
                    if word in words_dict:
                        # if the word is already present
                        # incrementing the counter for already existing words
                        words_dict[word] = words_dict.get(word)+1

                    else:
                        # first occurence of the word
                        words_dict[word] = 1 
    
   
    # converting the freqencies into list in the dictionary
    for key in words_dict:
        words_dict[key] = [words_dict.get(key)]
        #print(words_dict)
        

    #print(words_dict)
    # witing the dictionary into a dataframe
    
    
    # print("printing into df")
    words_df = pd.DataFrame(words_dict)
    words_df = words_df.T
    words_df.reset_index(inplace=True)
    words_df.columns = ["word","frequency"]
    words_df_sorted = words_df.sort_values(by="frequency", ascending = False)
    print(words_df_sorted)
    
                  
except Exception as e:
    print("Exception occured while reading the file "+str(e))
```
